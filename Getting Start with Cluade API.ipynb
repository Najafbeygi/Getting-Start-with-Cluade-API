```python
# Create an API client

from anthropic import Anthropic

client = Anthropic()

model = "claude-sonnet-4-0"
```

```python
#Make a request to the API

message = client.messages.create(
    model="claude-sonnet-4-0",
    max_tokens=100,       
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence."
        }
    ]
)

print(message.content[0].text)
```

```python
# does not have memorry, so we need to include the previous messages in the conversation
message = client.messages.create(
    model="claude-sonnet-4-0",
    max_tokens=100,       
    messages=[
        {
            "role": "user",
            "content": "write another sentence."
        } 
    ]
)

print(message.content[0].text)
```

```python
# To have a conversation

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)
    
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)
    
def chat(messages):
    messages = client.messages.create(
        model=model,
        max_tokens=100,
        messages=messages
    )
    return messages.content[0].text
```

```python
# Make a string list of messages
messages = []

# Add in the initial user question of "Define quantum computing in one sentence."
add_user_message(messages, "Define quantum computing in one sentence.")

print(messages)
# Pass the list of messages into 'Chat' to get an answer from the assistant
answer = chat(messages)

print(answer)
# Take the answer and add it as assistant message to the list of messages
add_assistant_message(messages, answer)

print(messages)
# Add in the users follow up question of "Write another sentence."
add_user_message(messages, "Write another sentence.")

#Call 'Chat' again with the list of message to get a final answer from the assistant
answer = chat(messages)
print(answer)
```

```python
# Make an initial list of messages
messages = []

# Use a 'while True loop' to run the chatbot forever
while True:
    #Get user input
    user_input = input("> ")
    print(">", user_input)
    
    #Add the user input as a message to the list of messages
    add_user_message(messages, user_input)
    #Call Cluade with the 'chat' function
    answer = chat(message)
    # Add generated text to the list of messages
    add_assistant_message(messages, answer)
    answer = chat(messages)
    # print the generated answer
    print("---------")
    print(answer)
    print("---------")
```
